```python
from tkinter import *
from tkinter import filedialog
import cv2
import numpy as np
import datetime as dt
import urllib.request
import ctypes
import time
```

```python
user32 = ctypes.windll.User32
def isLocked():
    return user32.GetForegroundWindow() == 0  
```

```python
def record():
    check = True
    timestr = time.strftime("%Y%m%d-%H%M%S")
    file_name = "output"+timestr+".avi"
    cap = cv2.VideoCapture(0)
    fourcc = cv2.VideoWriter_fourcc(*'MJPG')
    out = cv2.VideoWriter(file_name,fourcc,25.0,(640,480))

    while check:
        if isLocked():

            while cap.isOpened():
                
                ret,frame = cap.read()
                
                #frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
                if ret == True:
                    out.write(frame)
                else:
                    break
                
                if isLocked() == False:
                    check = False
                    break
        else:
        
            pass

    cap.release()
    out.release()
    cv2.destroyAllWindows()
    return file_name
```
